```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.platypus import Paragraph
from reportlab.lib.styles import ParagraphStyle
```

# Application Letter Generator
This script generates an application letter by placing building blocks of text and images on a pre-exported layout file (blank.png). To see the result, have a look at the file letter.pdf.

```python
# Register Fonts
pdfmetrics.registerFont(TTFont("Nunito", "Nunito-Medium.ttf"))
pdfmetrics.registerFont(TTFont("Nunito Bold", "Nunito-Bold.ttf"))
pdfmetrics.registerFont(TTFont("Nunito Sans", "Nunito-Sans.ttf"))

# Define Paragraph Styles
j_style = ParagraphStyle(
    name = "Justified",
    fontName = "Helvetica",
    fontSize = 12,
    textColor = (0.2, 0.2, 0.2),
    leading = 18,
    alignment = 4  # 0=left, 1=center, 2=right, 4=justify
)
```

```python
# Variable text buildings blocks
job_title = "[job description]"
company = "[company]"
date = "[date]"

# Static text building blocks
name = "[name]"
letter_text = f"""
Sehr geehrtes Team der {company},<br/><br/>

mit großem Interesse bewerbe ich mich bei Ihnen auf die ausgeschriebene Stelle als {job_title}. Im vergangenen Jahr habe ich mein Masterstudium im Bereich Data Science & Engineering erfolgreich abgeschlossen. Während dieses Studiums habe ich mir diverse Fähigkeiten in vielen Bereichen angeeignet, welche hervorragend auf Ihre Anforderungen passen. Weiters konnte ich im Rahmen von Projekten praktische Erfahrungen sammeln. Die Arbeit in diesem vielseitigen und spannenden Fachgebiet begeistert mich, da sie mir die Möglichkeit bietet, sowohl einen Beitrag für das Unternehmen zu leisten als auch im eigenen Interessensbereich neue Erkenntnisse zu gewinnen und innovative Technologien aktiv mitzugestalten.<br/><br/>

Nach Abschluss meines Master-Studiums plane ich nun, nach Wien zu ziehen, um meine berufliche Laufbahn bei einem innovativen Unternehmen wie dem Ihren fortzusetzen. Ihre Firma bietet aus meiner Sicht eine ausgezeichnete Umgebung, um mein Wissen und meine Erfahrung in diesem Gebiet weiter zu vertiefen und gleichzeitig einen wertvollen Beitrag zu leisten.<br/><br/>

Ich freue mich darauf, meine Fähigkeiten und Erfahrungen in einem persönlichen Gespräch weiter zu erläutern. Weitere Einblicke in meine Arbeitsweise finden Sie auch in meinem <a href="https://github.com/davidaigner98/datascience_portfolio"><u>Portfolio</u></a>.<br/><br/>

Mit freundlichen Grüßen,
"""
address = """
Wolfsbergstraße 7a
3373 Kemmelbach
Niederösterreich
"""
phone = "+43 680 4403804"
mail = "davidaigner98@gmail.com"
```

```python
# Calculate Positions for Building Blocks
w, h = A4
w_fac = w / 21
h_fac = h / 29.7

c1_x = (1.12 + 3.43) * w_fac
c2_x = 5.71 * w_fac

name_pos_y = (29.7 - 3.75) * h_fac
jobtitle_pos_y = (29.7 - 4.75) * h_fac
date_pos_y = (29.7 - 6.75) * h_fac
text_pos_y = (29.7 - 7.75) * h_fac

portrait_pos_y = (29.7 - 5.36) * h_fac
contact_pos_y = (29.7 - 6.75) * h_fac
address_pos_y = (29.7 - 8.0) * h_fac
phone_pos_y = (29.7 - 10.5) * h_fac
mail_pos_y = (29.7 - 12.3) * h_fac
```

```python
# Build PDF Document
c = canvas.Canvas(f"Application-Letter_{company}.pdf")

c.setTitle("Application Letter")

# Draw Background
w, h = A4
c.drawImage("blank.png", 0, 0, width=w, height=h)

# Add Building Blocks of 2nd Column
c.setFont('Nunito Sans', 48)
c.drawString(c2_x, name_pos_y, name)

jt = c.beginText()
jt.setTextOrigin(c2_x, jobtitle_pos_y)
jt.setFont('Nunito Sans', 14)
jt.setCharSpace(5.0)
jt.setFillColorRGB(0, 0, 0, 0.6)
jt.textLine(job_title.upper())
jt.setCharSpace(0)
jt.setFillColor("black")
c.drawText(jt)

c.setFont('Nunito Sans', 20)
c.drawString(c2_x + 2, date_pos_y, date)

p = Paragraph(letter_text, j_style)
w, h = p.wrap(380, 1000)
p.drawOn(c, c2_x, text_pos_y - h)

c.drawImage("signature.jpg", c2_x, text_pos_y - h - 2.53 * h_fac, width=7.49 * w_fac, height=2.53 * h_fac)

# Add Building Blocks of 1st Column
c.drawImage("portrait.jpg", c1_x - 3.43 * w_fac, portrait_pos_y, width=3.43 * w_fac, height=3.43 * h_fac)

c.setFont('Nunito Sans', 20)
c.drawRightString(c1_x, contact_pos_y, "Kontakt")
c.setFont('Nunito Sans', 14)
c.drawRightString(c1_x, address_pos_y, "Adresse")
c.setFont('Nunito', 8)
i = 0
for line in address.split("\n"):
    c.drawRightString(c1_x, address_pos_y - 12 * i, line)
    i += 1
c.setFont('Nunito Sans', 14)
c.drawRightString(c1_x, phone_pos_y, "Telefon")
c.setFont('Nunito', 8)
c.drawRightString(c1_x, phone_pos_y - 15, phone)
c.setFont('Nunito Sans', 14)
c.drawRightString(c1_x, mail_pos_y, "E-Mail")
c.setFont('Nunito', 8)
c.drawRightString(c1_x, mail_pos_y - 15, mail)

c.save()
```
